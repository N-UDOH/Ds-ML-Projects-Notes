### Day -3 | Sale Prediction from Existing  customer using LOGISTIC REGRESSION

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv(r"C:\Users\UDOH\Desktop\DigitalAd_dataset.csv")
```

```python
df.head(10)
```

```python
df.shape  #to get the shape of the dataset
```

```python
df.describe()
```

```python
df.info()
```

```python
  #To segregate the dataset into X(input/independent) and Y(output/dependent)
X = df[['Age', 'Salary']]  # Input features (age and salary)
Y = df['Status']# Target variable (status)
```

```python
#x.head()    # to confirm the segregation
Y.head()
```

```python
df.drop('feature',axis='columns')
Status
```

```python
#Segregating dataset into Train and Test
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
# Feature scaling: making all the features of X to be the same ie Age and Salary. DOnt touch Y
#we scale our data to make all the features contribute equally to the result
#Fit_transform-fit method is calculating the mean and variance of each of the
#features present in our data
#Transform- Transform method is transforming all the features using the mean and variance

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Training the model
model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
```

```python
age = int(input("Enter the age of the new customer: "))
salary = int(input("Enter the salary of the new customer: "))

# Create a list with the new customer's data
new_customer= [[age, salary]]

# Make predictions using the trained model
result = model.predict(sc.transform(new_customer))
print(result)
if result == 1:
    print("Customer will buy")
else:
    print("Customer Won't buy")
```

```python
# Prediction for all test data
y_pred = model.predict(x_test)
print(y_pred)
```

```python
#Evaluating model Using Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("Accuracy of the model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
